```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#read the CSV file into a DataFrame
heart_df = pd.read_csv('Heart Attack Data Set.csv')

#Display the first few rows of the DataFrame
heart_df.head()
```

```python
#Display the last few rows of the DataFrame
heart_df.tail()
```

```python
#Checking missing value
#'isnull()'Returns the null values per column
#'sum()' sums up the true values along with each column
heart_df.isnull().sum()
```

```python
#Generate descriptive statistics of the dataframe
heart_df.describe()
```

```python
#Calculate the correlation matrix for the dataframe
heart_df.corr()
```

```python
#Display the columns of the DataFrame
heart_df.columns
```

```python
#Set the figure size for the plot
plt.figure(figsize=(10,6))

# Create a count plot using seaborn
# plot the distribution of heart disease cases based on 'sex'
sns.countplot(data=heart_df,x='sex',hue='target',palette='viridis')

#Set the label for the x-axis
plt.xlabel("sex")

#Set the label for the y-axis
plt.ylabel("count")

#Set the title of the plot
plt.title("Heart Disease Counts by sex")

#Add a legend to the plot
plt.legend(title='Heart Disease',labels=['No Disease','Heart Disease'],loc='upper right')

#Display the plot
plt.show()

```

```python
#Set the figure size for the plot
plt.figure(figsize=(10,6))

#Create a scatter plot using seaborn
sns.scatterplot(data=heart_df, x='age', y='chol', hue='target', palette='dark',alpha=0.7)

# Set label for the x-axis
plt.xlabel("Age")

#Set the label for the y-axis
plt.ylabel("cholesterol level")

#Set the title of the plot
plt.title("Scatter plot of Age VS Cholesterol with heart disease")

#Add a legend to the plot
plt.legend(title='Heart Disease',labels=['No Disease','Heart Disease'],loc='upper right')

#Display the plot
plt.show()
```

```python
#compute the correlation matrix
correlation_matrix=heart_df.corr()

#Set the figure size
plt.figure(figsize=(12,10))

#Create a heatmap with a diverging color map
sns.heatmap(correlation_matrix,annot=True,cmap='coolwarm',fmt=".2f",linewidths=.5)

#Set the plot title
plt.title("Correlation Heatmap of Heart Dataset ")

#Display the plot
plt.show()
```
